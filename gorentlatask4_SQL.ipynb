2.WHERE

```python
 # Selecting title, edition,copyright from titles with copyright years after 2016
 pd.read_sql("""SELECT title, edition, copyright
                FROM titles 
                WHERE copyright > '2016'""",connection)
    
```

```python
 # All authors whose last name starts with D from Authors table
 pd.read_sql("""SELECT id, first, last
                FROM authors
                WHERE last LIKE 'D%'""",connection, index_col=['id'])
```

```python
 # All authors whose first name has b as the second character
 pd.read_sql("""SELECT id, first, last
                FROM authors
                WHERE first LIKE '_b%'""",
             connection, index_col=['id'])
```

3.ORDER BY

```python
 # Orderding titles ascending way
 pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
             connection)
```

```python
 # Sorting author names by last name first and later by first name
 pd.read_sql("""SELECT id, first, last
                FROM authors
                ORDER BY last,first""",
             connection,index_col=['id'])
```

```python
  # Sorting author names DESC last name and ASCE first name
 pd.read_sql("""SELECT id, first, last
                FROM authors
                ORDER BY last DESC,first ASC""",
             connection,index_col=['id'])
```

```python
# Fetching isbn, title, edition, copyright from titles where title ends with 'How to Program' and sorting by title ascending
pd.read_sql("""SELECT isbn, title, edition, copyright
               FROM titles
               WHERE title LIKE '%How to Program'
               ORDER BY title""", connection)
```

4.INNER JOIN   

```python
# Get data from authors and isbns tables
pd.read_sql("""SELECT first, last, isbn
               FROM authors
               INNER JOIN author_ISBN
               ON authors.id = author_ISBN.id
               ORDER BY last, first""", connection).head()
```
